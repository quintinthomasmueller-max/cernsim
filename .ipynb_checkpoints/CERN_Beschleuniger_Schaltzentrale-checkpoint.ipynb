## 🗺️ Der Protonen-Schienenweg am CERN
Bevor Protonen im **Large Hadron Collider (LHC)** bei nahezu Lichtgeschwindigkeit kollidieren, müssen sie eine präzise getaktete Kaskade von kleineren Vorbeschleunigern durchlaufen. Man kann sich dieses System wie ein hochpräzises Schienennetz vorstellen, bei dem Weichen im Nanosekundenbereich geschaltet werden, um die Teilchenpakete ("Bunches") in den nächsten Ring zu leiten:

1. **Wasserstoffquelle & LINAC 4 (Linear Accelerator 4)**: Der Startbahnhof. Wasserstoff-Ionen ($H^-$) werden erzeugt und linear auf eine Energie von **$160\text{ MeV}$** beschleunigt. Beim Eintritt in den nächsten Ring werden die Elektronen abgestreift - übrig bleiben reine Protonen.
2. **Proton Synchrotron Booster (PSB)**: Vier übereinanderliegende Ringe. Der Booster nimmt die Protonen auf und beschleunigt sie auf **$2\text{ GeV}$**.
3. **Proton Synchrotron (PS)**: Ein geschichtsträchtiger Ring mit einem Umfang von $628\text{ m}$. Er beschleunigt die Protonen auf **$26\text{ GeV}$** und formiert die präzisen Bunch-Strukturen.
4. **Super Proton Synchrotron (SPS)**: Ein $6.9\text{ km}$ langer Ring. Er dient als Hauptzubringer für den LHC und bringt die Protonen auf **$450\text{ GeV}$**.
5. **Large Hadron Collider (LHC)**: Der $26.7\text{ km}$ lange Hauptring. Über die beiden Transferlinien **TI 2** (mündet nahe ALICE) und **TI 8** (mündet nahe LHCb) werden zwei getrennte Protonenstrahlen eingespeist:
   - **Beam 1 (Blau)**: Läuft im Uhrzeigersinn (Clockwise).
   - **Beam 2 (Orange)**: Läuft gegen den Uhrzeigersinn (Counter-clockwise).
   - Im LHC werden die Strahlen auf die finale Kollisionsenergie von **$6.8\text{ TeV}$** pro Strahl gerampt (**$13.6\text{ TeV}$** Schwerpunktenergie) und an den vier Hauptkollisionspunkten (**ATLAS, CMS, ALICE, LHCb**) zur Kollision gebracht.

## 🧪 Physikalische Grundlagen

### 1. Relativistische Kinematik
Bei extrem hohen Energien wächst nicht mehr primär die Geschwindigkeit der Protonen (diese liegt bei $99.999999\%$ der Lichtgeschwindigkeit $c$), sondern ihre relativistische Masse bzw. Energie. Der Lorentz-Faktor $\gamma$ ist definiert als:
$$\gamma = \frac{E}{m_0 c^2} = \frac{1}{\sqrt{1 - \beta^2}}$$
wobei $\beta = v/c$.

### 2. Magnetische Ablenkung (Dipolfeld)
Um die geladenen Protonen auf ihrer Kreisbahn mit Radius $R$ zu halten, müssen die supraleitenden Dipolmagnete ein Magnetfeld $B$ erzeugen, das synchron mit der Energie ansteigt (Ramping):
$$B [\text{Tesla}] = \frac{p [\text{GeV}/c]}{0.299792 \cdot R [\text{m}]}$$
Im LHC beträgt der effektive Biegeradius der Dipolmagnete $R \approx 2804\text{ m}$. Bei einer Energie von $6.8\text{ TeV}$ ($6800\text{ GeV}$) entspricht dies einem Magnetfeld von **$\approx 8.09\text{ Tesla}$**!


```python
# Importieren der CERN-Hilfsbibliothek und Standardpakete
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Pfad für cern_utils.py hinzufügen
sys.path.append(os.path.abspath('../scripts'))
try:
    import cern_utils as cu
    print('✅ CERN-Hilfsbibliothek erfolgreich geladen!')
except ImportError:
    print('⚠️ cern_utils.py konnte nicht importiert werden. Verwende Standard-Hilfsfunktionen.')

# ------------------------------------------------------------------------------
# Physik-Modellierung des Beschleunigerkomplexes
# ------------------------------------------------------------------------------
class CERNComplexSimulator:
    def __init__(self):
        # Radien der Beschleuniger (in Metern)
        self.radii = {
            'LINAC4': 0.0,      # Linear
            'PSB': 25.0,        # R ~ 25 m
            'PS': 100.0,        # R ~ 100 m (Umfang 628m)
            'SPS': 1100.0,      # R ~ 1100 m (Umfang 6.9km)
            'LHC': 4243.0       # R ~ 4243 m (Biegeradius der Magnete R = 2804 m)
        }
        # Magnetischer Biegeradius des LHC
        self.LHC_bending_radius = 2803.95
        
        # Maximalenergien der Stufen
        self.max_energies = {
            'LINAC4': 0.16,     # 160 MeV
            'PSB': 2.0,         # 2 GeV
            'PS': 26.0,         # 26 GeV
            'SPS': 450.0,       # 450 GeV
            'LHC': 6800.0       # 6.8 TeV (6800 GeV)
        }
        
        # Konstanten
        self.c = 299792458.0 # m/s
        self.mp = 0.93827208 # Protonenmasse in GeV/c^2
        
    def berechne_relativistische_werte(self, energie_gev):
        """Berechnet Lorentz-Faktor, Beta (v/c) und Geschwindigkeit in m/s."""
        gamma = energie_gev / self.mp
        beta = np.sqrt(1.0 - 1.0 / (gamma**2)) if gamma > 1.0 else 0.0
        geschwindigkeit = beta * self.c
        return gamma, beta, geschwindigkeit
        
    def berechne_magnetfeld_lhc(self, energie_gev):
        """Berechnet die benötigte Feldstärke der LHC-Dipole in Tesla."""
        # p = q * B * R => B = p / (0.3 * R) bei e+ Ladung
        impuls = np.sqrt(np.maximum(energie_gev**2 - self.mp**2, 0.0))
        B = impuls / (0.299792458 * self.LHC_bending_radius)
        return B
        
    def generiere_higgs_zerfall(self):
        """Simuliert den Zerfall H -> gamma gamma (zwei Photonen)."""
        # Higgs-Masse ~ 125 GeV
        higgs_mass = 125.0
        # Photonenimpulse im Ruhesystem des Higgs: E = m/2 = 62.5 GeV
        # Zufällige Raumrichtung
        theta = np.arccos(np.random.uniform(-1, 1))
        phi = np.random.uniform(0, 2 * np.pi)
        
        p_x = 62.5 * np.sin(theta) * np.cos(phi)
        p_y = 62.5 * np.sin(theta) * np.sin(phi)
        p_z = 62.5 * np.cos(theta)
        
        # Photon 1 & 2 (entgegengesetzt)
        photon1 = {'px': p_x, 'py': p_y, 'pz': p_z, 'E': 62.5}
        photon2 = {'px': -p_x, 'py': -p_y, 'pz': -p_z, 'E': 62.5}
        return photon1, photon2

sim = CERNComplexSimulator()
print('⚙️ Physik-Engine initialisiert.')
```

```python
display(HTML('''
<div id="cern-stellwerk-container">
    <style>
        #cern-stellwerk-container {
            background: #0d1117;
            color: #c9d1d9;
            font-family: 'SF Pro Display', -apple-system, 'Segoe UI', Roboto, Helvetica, sans-serif;
            border-radius: 16px;
            padding: 24px;
            box-shadow: 0 8px 32px 0 rgba(0, 0, 0, 0.5);
            border: 1px solid #30363d;
            max-width: 1100px;
            margin: 0 auto;
            user-select: none;
        }
        
        .cern-header {
            display: flex;
            justify-content: space-between;
            align-items: center;
            border-bottom: 2px solid #21262d;
            padding-bottom: 12px;
            margin-bottom: 20px;
        }
        
        .cern-logo-text {
            font-size: 22px;
            font-weight: 800;
            color: #58a6ff;
            letter-spacing: 1px;
            display: flex;
            align-items: center;
            gap: 10px;
        }
        
        .cern-badge {
            background: rgba(88, 166, 255, 0.15);
            color: #58a6ff;
            font-size: 11px;
            padding: 4px 8px;
            border-radius: 12px;
            border: 1px solid rgba(88, 166, 255, 0.3);
            text-transform: uppercase;
        }

        .cern-grid {
            display: grid;
            grid-template-columns: 1fr 320px;
            gap: 24px;
        }

        @media (max-width: 900px) {
            .cern-grid {
                grid-template-columns: 1fr;
            }
        }

        .control-panel {
            background: #161b22;
            border-radius: 12px;
            border: 1px solid #30363d;
            padding: 16px;
            display: flex;
            flex-direction: column;
            gap: 16px;
        }

        .panel-title {
            font-size: 14px;
            text-transform: uppercase;
            letter-spacing: 1px;
            color: #8b949e;
            border-bottom: 1px solid #30363d;
            padding-bottom: 8px;
            margin-bottom: 8px;
            display: flex;
            justify-content: space-between;
            align-items: center;
        }

        .stellwerk-svg-wrapper {
            background: #090d13;
            border-radius: 12px;
            border: 1px solid #21262d;
            position: relative;
            overflow: hidden;
            display: flex;
            justify-content: center;
            align-items: center;
            height: 480px;
        }

        .interactive-btn {
            background: #21262d;
            color: #c9d1d9;
            border: 1px solid #30363d;
            padding: 10px 16px;
            border-radius: 6px;
            cursor: pointer;
            font-size: 13px;
            font-weight: 600;
            transition: all 0.2s ease;
            display: flex;
            align-items: center;
            justify-content: center;
            gap: 8px;
        }

        .interactive-btn:hover {
            background: #30363d;
            border-color: #8b949e;
            color: #f0f6fc;
        }

        .btn-active {
            background: rgba(88, 166, 255, 0.15);
            border-color: #58a6ff;
            color: #58a6ff;
            box-shadow: 0 0 10px rgba(88, 166, 255, 0.2);
        }

        .btn-active:hover {
            background: rgba(88, 166, 255, 0.25);
            border-color: #58a6ff;
        }

        .btn-danger {
            background: rgba(248, 81, 73, 0.1);
            border-color: rgba(248, 81, 73, 0.4);
            color: #f85149;
        }

        .btn-danger:hover {
            background: rgba(248, 81, 73, 0.2);
            border-color: #f85149;
            box-shadow: 0 0 12px rgba(248, 81, 73, 0.3);
        }

        .btn-disabled {
            opacity: 0.4;
            cursor: not-allowed;
            pointer-events: none;
        }

        /* SVG styling */
        .svg-beam-path {
            stroke: #21262d;
            stroke-width: 3;
            fill: none;
            transition: stroke 0.3s ease;
        }

        .svg-beam-active {
            stroke: #58a6ff;
            filter: drop-shadow(0 0 6px #58a6ff);
        }

        .svg-beam-active-b2 {
            stroke: #ff7f0e;
            filter: drop-shadow(0 0 6px #ff7f0e);
        }

        .svg-beam-lhc-b1 {
            stroke: rgba(88, 166, 255, 0.2);
            stroke-width: 4;
        }

        .svg-beam-lhc-b2 {
            stroke: rgba(255, 127, 14, 0.2);
            stroke-width: 4;
        }

        .svg-beam-lhc-b1.active {
            stroke: #58a6ff;
            filter: drop-shadow(0 0 8px #58a6ff);
        }

        .svg-beam-lhc-b2.active {
            stroke: #ff7f0e;
            filter: drop-shadow(0 0 8px #ff7f0e);
        }

        .svg-node {
            fill: #161b22;
            stroke: #30363d;
            stroke-width: 2;
            transition: all 0.3s ease;
        }

        .svg-node-active {
            stroke: #58a6ff;
            fill: rgba(88, 166, 255, 0.1);
            filter: drop-shadow(0 0 8px rgba(88, 166, 255, 0.5));
        }

        .svg-node-active-red {
            stroke: #f85149;
            fill: rgba(248, 81, 73, 0.1);
            filter: drop-shadow(0 0 8px rgba(248, 81, 73, 0.5));
        }

        .svg-label {
            font-size: 10px;
            fill: #8b949e;
            font-family: monospace;
            text-anchor: middle;
        }

        .svg-label-active {
            fill: #f0f6fc;
            font-weight: bold;
        }

        /* Running proton animation */
        .proton-b1 {
            fill: #58a6ff;
            filter: drop-shadow(0 0 4px #58a6ff);
        }

        .proton-b2 {
            fill: #ff7f0e;
            filter: drop-shadow(0 0 4px #ff7f0e);
        }

        /* Readout display elements */
        .readout-grid {
            display: grid;
            grid-template-columns: 1fr 1fr;
            gap: 10px;
        }

        .readout-box {
            background: #0d1117;
            border-radius: 6px;
            border: 1px solid #21262d;
            padding: 8px 12px;
            display: flex;
            flex-direction: column;
        }

        .readout-label {
            font-size: 10px;
            color: #8b949e;
            text-transform: uppercase;
        }

        .readout-value {
            font-size: 16px;
            font-weight: 700;
            color: #f0f6fc;
            font-family: 'Courier New', Courier, monospace;
        }

        /* Ramping Progress Bar */
        .progress-bar-wrapper {
            background: #21262d;
            border-radius: 4px;
            height: 8px;
            overflow: hidden;
            width: 100%;
        }

        .progress-bar-fill {
            background: linear-gradient(90deg, #58a6ff, #17becf);
            height: 100%;
            width: 0%;
            transition: width 0.1s linear;
        }

        /* Detector View & Event Display */
        .event-display-wrapper {
            display: flex;
            flex-direction: column;
            gap: 10px;
        }

        .event-canvas {
            background: #090d13;
            border: 1px solid #21262d;
            border-radius: 8px;
            width: 100%;
            height: 180px;
        }

        .detector-selector {
            display: grid;
            grid-template-columns: repeat(4, 1fr);
            gap: 4px;
        }

        .detector-tab {
            background: #21262d;
            border: 1px solid #30363d;
            padding: 6px;
            font-size: 11px;
            color: #8b949e;
            border-radius: 4px;
            cursor: pointer;
            text-align: center;
            font-weight: bold;
        }

        .detector-tab.active {
            background: rgba(88, 166, 255, 0.15);
            border-color: #58a6ff;
            color: #58a6ff;
        }

        /* Live histogram overlay */
        .histogram-container {
            background: #090d13;
            border: 1px solid #21262d;
            border-radius: 8px;
            height: 120px;
            position: relative;
            padding: 8px;
        }

        .info-status {
            font-size: 11px;
            color: #8b949e;
            display: flex;
            align-items: center;
            gap: 6px;
        }

        .status-dot {
            width: 8px;
            height: 8px;
            border-radius: 50%;
            background: #8b949e;
            display: inline-block;
        }

        .status-dot.active {
            background: #2ea44f;
            box-shadow: 0 0 8px #2ea44f;
            animation: pulse 1.5s infinite;
        }

        .status-dot.danger {
            background: #f85149;
            box-shadow: 0 0 8px #f85149;
            animation: pulse-red 1.5s infinite;
        }

        @keyframes pulse {
            0% { transform: scale(0.9); opacity: 0.6; }
            50% { transform: scale(1.2); opacity: 1; }
            100% { transform: scale(0.9); opacity: 0.6; }
        }

        @keyframes pulse-red {
            0% { transform: scale(0.9); opacity: 0.6; }
            50% { transform: scale(1.2); opacity: 1; }
            100% { transform: scale(0.9); opacity: 0.6; }
        }
    </style>

    <div class="cern-header">
        <div class="cern-logo-text">
            <span>⚛️ CERN CONTROL ROOM</span>
            <span class="cern-badge">Schienenschaltzentrale v2.0</span>
        </div>
        <div class="info-status">
            <span class="status-dot" id="system-status-dot"></span>
            <span id="system-status-text">OFFLINE</span>
        </div>
    </div>

    <div class="cern-grid">
        <!-- LINKER TEIL: Stellwerkkarte (SVG) -->
        <div class="stellwerk-svg-wrapper">
            <svg id="stellwerk-svg" width="680" height="460" viewBox="0 0 680 460">
                <!-- Defs for glowing effects -->
                <defs>
                    <filter id="glow-cyan" x="-20%" y="-20%" width="140%" height="140%">
                        <feGaussianBlur stdDeviation="5" result="blur" />
                        <feMerge>
                            <feMergeNode in="blur" />
                            <feMergeNode in="SourceGraphic" />
                        </feMerge>
                    </filter>
                    <filter id="glow-orange" x="-20%" y="-20%" width="140%" height="140%">
                        <feGaussianBlur stdDeviation="5" result="blur" />
                        <feMerge>
                            <feMergeNode in="blur" />
                            <feMergeNode in="SourceGraphic" />
                        </feMerge>
                    </filter>
                </defs>

                <!-- Grid Background lines (Sci-fi decoration) -->
                <line x1="0" y1="230" x2="680" y2="230" stroke="#161b22" stroke-width="1" />
                <line x1="340" y1="0" x2="340" y2="460" stroke="#161b22" stroke-width="1" />

                <!-- == WEGE (Gleise / Pfade) == -->
                <!-- 1. Linac 4 (Gerader Einlasspfad) -->
                <path id="path-linac4" d="M 40,360 L 140,360" class="svg-beam-path" />
                
                <!-- 2. PSB Ring -->
                <circle id="path-psb" cx="180" cy="360" r="30" class="svg-beam-path" />
                
                <!-- 3. Transfer PSB -> PS -->
                <path id="path-psb-ps" d="M 195,334 Q 210,300 230,300" class="svg-beam-path" />
                
                <!-- 4. PS Ring -->
                <circle id="path-ps" cx="280" cy="300" r="55" class="svg-beam-path" />
                
                <!-- 5. Transfer PS -> SPS -->
                <path id="path-ps-sps" d="M 330,280 Q 360,250 370,220" class="svg-beam-path" />
                
                <!-- 6. SPS Ring -->
                <circle id="path-sps" cx="440" cy="180" r="95" class="svg-beam-path" />
                
                <!-- 7. LHC Ring (Riesiger Hauptring) -->
                <!-- Center cx=350, cy=230, r=210 -->
                <circle id="path-lhc-b1" cx="350" cy="230" r="210" class="svg-beam-path svg-beam-lhc-b1" />
                <circle id="path-lhc-b2" cx="350" cy="230" r="210" class="svg-beam-path svg-beam-lhc-b2" />

                <!-- 8. TI 2 (Transfer line from SPS to LHC near ALICE) -->
                <path id="path-ti2" d="M 345,185 Q 240,185 201,311" class="svg-beam-path" />

                <!-- 9. TI 8 (Transfer line from SPS to LHC near LHCb) -->
                <path id="path-ti8" d="M 525,220 Q 550,260 498,374" class="svg-beam-path" />

                <!-- == BAHNHOFE / EXPERIMENTE (Kollisionspunkte & Beschleuniger-Nodes) == -->
                <!-- LINAC 4 Startbahnhof -->
                <circle id="node-linac4" cx="40" cy="360" r="6" class="svg-node" />
                <text x="40" y="380" class="svg-label">LINAC 4</text>

                <!-- PSB Node -->
                <circle id="node-psb" cx="180" cy="360" r="8" class="svg-node" />
                <text x="180" y="405" class="svg-label">PSB</text>

                <!-- PS Node -->
                <circle id="node-ps" cx="280" cy="300" r="9" class="svg-node" />
                <text x="280" y="375" class="svg-label">PS</text>

                <!-- SPS Node -->
                <circle id="node-sps" cx="440" cy="180" r="12" class="svg-node" />
                <text x="440" y="295" class="svg-label">SPS</text>

                <!-- LHC DETECTORS (Point 1, 2, 5, 8) -->
                <!-- Point 1: ATLAS (Unten, 350, 440) -->
                <circle id="det-atlas" cx="350" cy="440" r="16" class="svg-node" />
                <text x="350" y="470" class="svg-label" style="font-weight:bold;fill:#fff;">ATLAS (IP1)</text>

                <!-- Point 2: ALICE (Links unten, 201.5, 378.4) -->
                <circle id="det-alice" cx="201.5" cy="378.4" r="14" class="svg-node" />
                <text x="160" y="375" class="svg-label" style="font-weight:bold;fill:#fff;">ALICE (IP2)</text>

                <!-- Point 5: CMS (Oben, 350, 20) -->
                <circle id="det-cms" cx="350" cy="20" r="16" class="svg-node" />
                <text x="350" y="-5" class="svg-label" style="font-weight:bold;fill:#fff;">CMS (IP5)</text>

                <!-- Point 8: LHCb (Rechts unten, 498.5, 378.4) -->
                <circle id="det-lhcb" cx="498.5" cy="378.4" r="14" class="svg-node" />
                <text x="540" y="375" class="svg-label" style="font-weight:bold;fill:#fff;">LHCb (IP8)</text>

                <!-- Proton particle beam circles (Beams running on tracks) -->
                <circle id="beam-particle-linac" cx="-50" cy="-50" r="4" class="proton-b1" style="display:none;" />
                <circle id="beam-particle-psb" cx="-50" cy="-50" r="4" class="proton-b1" style="display:none;" />
                <circle id="beam-particle-ps" cx="-50" cy="-50" r="4.5" class="proton-b1" style="display:none;" />
                <circle id="beam-particle-sps" cx="-50" cy="-50" r="5" class="proton-b1" style="display:none;" />
                
                <!-- LHC Proton packages (clockwise and counterclockwise) -->
                <circle id="lhc-b1-particle" cx="-50" cy="-50" r="5" class="proton-b1" style="display:none;" />
                <circle id="lhc-b2-particle" cx="-50" cy="-50" r="5" class="proton-b2" style="display:none;" />
            </svg>
        </div>

        <!-- RECHTER TEIL: Stellpult-Kontrollen & Messwerte -->
        <div class="control-panel">
            <div>
                <div class="panel-title">📡 BAHNSTATION-BEDIENUNG</div>
                <div style="display:flex; flex-direction:column; gap:8px;">
                    <button class="interactive-btn" id="btn-inject-linac">📥 1. Linac4 / PSB laden</button>
                    <button class="interactive-btn btn-disabled" id="btn-transfer-ps">🛤️ 2. Transfer PSB ➔ PS</button>
                    <button class="interactive-btn btn-disabled" id="btn-transfer-sps">🛤️ 3. Transfer PS ➔ SPS</button>
                    <div style="display:grid; grid-template-columns:1fr 1fr; gap:6px;">
                        <button class="interactive-btn btn-disabled" id="btn-inject-lhc-b1">🔄 LHC B1 (CW)</button>
                        <button class="interactive-btn btn-disabled" id="btn-inject-lhc-b2">🔄 LHC B2 (CCW)</button>
                    </div>
                </div>
            </div>

            <div>
                <div class="panel-title">⚡ LHC BETRIEBSART</div>
                <div style="display:flex; flex-direction:column; gap:8px;">
                    <button class="interactive-btn btn-disabled" id="btn-ramp-lhc">🚀 Beschleunigen (Ramping)</button>
                    <div class="progress-bar-wrapper">
                        <div class="progress-bar-fill" id="lhc-progress-bar"></div>
                    </div>
                    <button class="interactive-btn btn-danger btn-disabled" id="btn-collide">💥 Strahlkollision auslösen!</button>
                </div>
            </div>

            <div>
                <div class="panel-title">📊 KONTROLLRAUM DATEN</div>
                <div class="readout-grid">
                    <div class="readout-box">
                        <span class="readout-label">LHC Energie (beam)</span>
                        <span class="readout-value" id="val-energy">0.00 GeV</span>
                    </div>
                    <div class="readout-box">
                        <span class="readout-label">Magnetfeld B</span>
                        <span class="readout-value" id="val-field">0.000 T</span>
                    </div>
                    <div class="readout-box">
                        <span class="readout-label">Luminosität</span>
                        <span class="readout-value" id="val-lumi">0.00 Hz/cm²</span>
                    </div>
                    <div class="readout-box">
                        <span class="readout-label">Lorentz Gamma</span>
                        <span class="readout-value" id="val-gamma">1.00</span>
                    </div>
                </div>
            </div>
        </div>
    </div>

    <!-- UNTERER TEIL: Live Detektor Event Display & Analyse -->
    <div style="margin-top: 24px; display: grid; grid-template-columns: 1fr 1fr; gap: 20px;">
        <!-- Event-Display -->
        <div class="event-display-wrapper">
            <div class="panel-title">📸 DETEKTOR EVENT DISPLAY</div>
            <div class="detector-selector">
                <div class="detector-tab active" id="tab-atlas">ATLAS</div>
                <div class="detector-tab" id="tab-cms">CMS</div>
                <div class="detector-tab" id="tab-alice">ALICE</div>
                <div class="detector-tab" id="tab-lhcb">LHCb</div>
            </div>
            <canvas class="event-canvas" id="canvas-event"></canvas>
        </div>

        <!-- Analyse Histogramm -->
        <div class="event-display-wrapper">
            <div class="panel-title">📊 MASSENSPEKTRUM REKONSTRUKTION</div>
            <div style="display:flex; justify-content:space-between; align-items:center; font-size:11px; margin-bottom:-4px;">
                <span id="collision-counter" style="color:#58a6ff;">Kollisionen: 0</span>
                <span style="color:#8b949e;">Massebereich: 50 - 150 GeV/c²</span>
            </div>
            <div class="histogram-container">
                <canvas id="canvas-histogram" style="width:100%; height:100%;"></canvas>
            </div>
        </div>
    </div>
</div>

<script>
(function() {
    // CERN Accelerator physics constants
    const proton_mass = 0.938272; // GeV
    const c_speed = 299792458; // m/s
    const bending_radius = 2803.95; // m

    // State Variables
    let system_state = "OFFLINE"; // OFFLINE, LINAC_LOADED, PS_LOADED, SPS_LOADED, LHC_B1, LHC_B2, LHC_BOTH, LHC_RAMPING, LHC_STABLE
    let lhc_energy = 450.0; // Starting from injection energy (GeV)
    let collision_count = 0;
    
    // Arrays for simulated collision mass data
    let accumulated_mass_data = [];
    
    // SVG Elements
    const svg = document.getElementById("stellwerk-svg");
    const path_linac4 = document.getElementById("path-linac4");
    const path_psb = document.getElementById("path-psb");
    const path_psb_ps = document.getElementById("path-psb-ps");
    const path_ps = document.getElementById("path-ps");
    const path_ps_sps = document.getElementById("path-ps-sps");
    const path_sps = document.getElementById("path-sps");
    const path_ti2 = document.getElementById("path-ti2");
    const path_ti8 = document.getElementById("path-ti8");
    const path_lhc_b1 = document.getElementById("path-lhc-b1");
    const path_lhc_b2 = document.getElementById("path-lhc-b2");

    // Nodes
    const node_linac = document.getElementById("node-linac4");
    const node_psb = document.getElementById("node-psb");
    const node_ps = document.getElementById("node-ps");
    const node_sps = document.getElementById("node-sps");
    const det_atlas = document.getElementById("det-atlas");
    const det_cms = document.getElementById("det-cms");
    const det_alice = document.getElementById("det-alice");
    const det_lhcb = document.getElementById("det-lhcb");

    // Beam particle animators
    const p_linac = document.getElementById("beam-particle-linac");
    const p_psb = document.getElementById("beam-particle-psb");
    const p_ps = document.getElementById("beam-particle-ps");
    const p_sps = document.getElementById("beam-particle-sps");
    const p_lhc_b1 = document.getElementById("lhc-b1-particle");
    const p_lhc_b2 = document.getElementById("lhc-b2-particle");

    // Buttons
    const btn_inject_linac = document.getElementById("btn-inject-linac");
    const btn_transfer_ps = document.getElementById("btn-transfer-ps");
    const btn_transfer_sps = document.getElementById("btn-transfer-sps");
    const btn_inject_lhc_b1 = document.getElementById("btn-inject-lhc-b1");
    const btn_inject_lhc_b2 = document.getElementById("btn-inject-lhc-b2");
    const btn_ramp_lhc = document.getElementById("btn-ramp-lhc");
    const btn_collide = document.getElementById("btn-collide");

    // Readouts
    const val_energy = document.getElementById("val-energy");
    const val_field = document.getElementById("val-field");
    const val_lumi = document.getElementById("val-lumi");
    const val_gamma = document.getElementById("val-gamma");
    const status_dot = document.getElementById("system-status-dot");
    const status_text = document.getElementById("system-status-text");
    const lhc_progress_bar = document.getElementById("lhc-progress-bar");
    const collision_counter_text = document.getElementById("collision-counter");

    // Canvas Elements
    const canvas_event = document.getElementById("canvas-event");
    const ctx_event = canvas_event.getContext("2d");
    const canvas_hist = document.getElementById("canvas-histogram");
    const ctx_hist = canvas_hist.getContext("2d");

    // Selected Detector
    let selected_detector = "ATLAS";

    // Setup Detector Selection tabs
    const detector_tabs = {
        "ATLAS": document.getElementById("tab-atlas"),
        "CMS": document.getElementById("tab-cms"),
        "ALICE": document.getElementById("tab-alice"),
        "LHCb": document.getElementById("tab-lhcb")
    };

    Object.keys(detector_tabs).forEach(det => {
        detector_tabs[det].addEventListener("click", () => {
            Object.values(detector_tabs).forEach(tab => tab.classList.remove("active"));
            detector_tabs[det].classList.add("active");
            selected_detector = det;
            drawStaticDetectorBackground();
        });
    });

    // Event Loop Timing / Animations variables
    let animators = {};

    function updateSystemStatus(state, desc, type="active") {
        system_state = state;
        status_text.innerText = desc;
        status_dot.className = "status-dot " + type;
    }

    // Set Initial state
    updateSystemStatus("OFFLINE", "INJEKTOREN LEER", "danger");
    drawStaticDetectorBackground();
    drawHistogram();

    // Utility function: Get coordinate at path length
    function getPointOnPath(pathElement, progress) {
        // progress is 0 to 1
        const totalLength = pathElement.getTotalLength();
        return pathElement.getPointAtLength(progress * totalLength);
    }

    // 1. INJECTION: LINAC4 -> PSB
    btn_inject_linac.addEventListener("click", () => {
        btn_inject_linac.classList.add("btn-active");
        node_linac.classList.add("svg-node-active");
        
        path_linac4.classList.add("svg-beam-active");
        path_psb.classList.add("svg-beam-active");
        node_psb.classList.add("svg-node-active");

        p_linac.style.display = "block";
        p_psb.style.display = "block";

        let start = null;
        function anim(timestamp) {
            if (!start) start = timestamp;
            let elapsed = timestamp - start;
            
            // Move along LINAC4
            let linac_progress = (elapsed % 1500) / 1500;
            let pt_linac = getPointOnPath(path_linac4, linac_progress);
            p_linac.setAttribute("cx", pt_linac.x);
            p_linac.setAttribute("cy", pt_linac.y);

            // Orbit in PSB Ring
            let psb_progress = (elapsed % 800) / 800;
            let pt_psb = getPointOnPath(path_psb, psb_progress);
            p_psb.setAttribute("cx", pt_psb.x);
            p_psb.setAttribute("cy", pt_psb.y);

            animators.inject = requestAnimationFrame(anim);
        }
        
        if (animators.inject) cancelAnimationFrame(animators.inject);
        animators.inject = requestAnimationFrame(anim);

        updateSystemStatus("LINAC_LOADED", "BEAM IN PSB (2 GeV)");
        
        btn_transfer_ps.classList.remove("btn-disabled");
    });

    // 2. TRANSFER PSB -> PS
    btn_transfer_ps.addEventListener("click", () => {
        btn_transfer_ps.classList.add("btn-active");
        path_psb_ps.classList.add("svg-beam-active");
        path_ps.classList.add("svg-beam-active");
        node_ps.classList.add("svg-node-active");
        p_ps.style.display = "block";

        let start = null;
        function anim(timestamp) {
            if (!start) start = timestamp;
            let elapsed = timestamp - start;

            // Move around PS Ring
            let ps_progress = (elapsed % 1000) / 1000;
            let pt_ps = getPointOnPath(path_ps, ps_progress);
            p_ps.setAttribute("cx", pt_ps.x);
            p_ps.setAttribute("cy", pt_ps.y);

            animators.ps = requestAnimationFrame(anim);
        }
        if (animators.ps) cancelAnimationFrame(animators.ps);
        animators.ps = requestAnimationFrame(anim);

        updateSystemStatus("PS_LOADED", "BEAM IN PS (26 GeV)");
        btn_transfer_sps.classList.remove("btn-disabled");
    });

    // 3. TRANSFER PS -> SPS
    btn_transfer_sps.addEventListener("click", () => {
        btn_transfer_sps.classList.add("btn-active");
        path_ps_sps.classList.add("svg-beam-active");
        path_sps.classList.add("svg-beam-active");
        node_sps.classList.add("svg-node-active");
        p_sps.style.display = "block";

        let start = null;
        function anim(timestamp) {
            if (!start) start = timestamp;
            let elapsed = timestamp - start;

            // Move around SPS Ring
            let sps_progress = (elapsed % 1200) / 1200;
            let pt_sps = getPointOnPath(path_sps, sps_progress);
            p_sps.setAttribute("cx", pt_sps.x);
            p_sps.setAttribute("cy", pt_sps.y);

            animators.sps = requestAnimationFrame(anim);
        }
        if (animators.sps) cancelAnimationFrame(animators.sps);
        animators.sps = requestAnimationFrame(anim);

        updateSystemStatus("SPS_LOADED", "BEAM IN SPS (450 GeV)");
        btn_inject_lhc_b1.classList.remove("btn-disabled");
        btn_inject_lhc_b2.classList.remove("btn-disabled");
    });

    // 4. INJECT LHC BEAM 1 (Clockwise - Cyan)
    btn_inject_lhc_b1.addEventListener("click", () => {
        btn_inject_lhc_b1.classList.add("btn-active");
        path_ti2.classList.add("svg-beam-active");
        path_lhc_b1.classList.add("active");
        p_lhc_b1.style.display = "block";

        let start = null;
        function anim(timestamp) {
            if (!start) start = timestamp;
            let elapsed = timestamp - start;

            // Orbit Clockwise
            let lhc_progress = (elapsed % 2500) / 2500;
            let pt = getPointOnPath(path_lhc_b1, lhc_progress);
            p_lhc_b1.setAttribute("cx", pt.x);
            p_lhc_b1.setAttribute("cy", pt.y);

            animators.lhc_b1 = requestAnimationFrame(anim);
        }
        if (animators.lhc_b1) cancelAnimationFrame(animators.lhc_b1);
        animators.lhc_b1 = requestAnimationFrame(anim);

        if (system_state === "LHC_B2" || system_state === "LHC_BOTH") {
            updateSystemStatus("LHC_BOTH", "BEAMS IN LHC (450 GeV)");
            btn_ramp_lhc.classList.remove("btn-disabled");
        } else {
            updateSystemStatus("LHC_B1", "LHC BEAM 1 INJECTED");
        }
        updateDashboardPhysics(450.0);
    });

    // 5. INJECT LHC BEAM 2 (Counter-Clockwise - Orange)
    btn_inject_lhc_b2.addEventListener("click", () => {
        btn_inject_lhc_b2.classList.add("btn-active");
        path_ti8.classList.add("svg-beam-active-b2");
        path_lhc_b2.classList.add("active");
        p_lhc_b2.style.display = "block";

        let start = null;
        function anim(timestamp) {
            if (!start) start = timestamp;
            let elapsed = timestamp - start;

            // Orbit Counter-Clockwise (1 - progress)
            let lhc_progress = 1 - ((elapsed % 2500) / 2500);
            let pt = getPointOnPath(path_lhc_b2, lhc_progress);
            p_lhc_b2.setAttribute("cx", pt.x);
            p_lhc_b2.setAttribute("cy", pt.y);

            animators.lhc_b2 = requestAnimationFrame(anim);
        }
        if (animators.lhc_b2) cancelAnimationFrame(animators.lhc_b2);
        animators.lhc_b2 = requestAnimationFrame(anim);

        if (system_state === "LHC_B1" || system_state === "LHC_BOTH") {
            updateSystemStatus("LHC_BOTH", "BEAMS IN LHC (450 GeV)");
            btn_ramp_lhc.classList.remove("btn-disabled");
        } else {
            updateSystemStatus("LHC_B2", "LHC BEAM 2 INJECTED");
        }
        updateDashboardPhysics(450.0);
    });

    // 6. RAMP ENERGY
    btn_ramp_lhc.addEventListener("click", () => {
        btn_ramp_lhc.classList.add("btn-active");
        updateSystemStatus("LHC_RAMPING", "RAMPING ENERGY (450 GeV -> 6.8 TeV)");
        
        let start_time = null;
        const duration = 4000; // 4 seconds ramp
        const start_energy = 450.0;
        const target_energy = 6800.0;

        function ramp(timestamp) {
            if (!start_time) start_time = timestamp;
            let elapsed = timestamp - start_time;
            let progress = Math.min(elapsed / duration, 1.0);
            
            lhc_energy = start_energy + progress * (target_energy - start_energy);
            updateDashboardPhysics(lhc_energy);

            lhc_progress_bar.style.width = (progress * 100) + "%";

            // Increase particle visual speed proportional to gamma (just for effect)
            // Let's cancel and re-run LHC orbit animations with faster speeds
            let speed_b1 = 2500 - (progress * 1500);
            let speed_b2 = 2500 - (progress * 1500);
            
            cancelAnimationFrame(animators.lhc_b1);
            animators.lhc_b1 = requestAnimationFrame((ts) => {
                let lhc_progress = ((ts - start_time) % speed_b1) / speed_b1;
                let pt = getPointOnPath(path_lhc_b1, lhc_progress);
                p_lhc_b1.setAttribute("cx", pt.x);
                p_lhc_b1.setAttribute("cy", pt.y);
                
                // Keep orbiting
                function orbit(t) {
                    let p = (t % speed_b1) / speed_b1;
                    let pt = getPointOnPath(path_lhc_b1, p);
                    p_lhc_b1.setAttribute("cx", pt.x);
                    p_lhc_b1.setAttribute("cy", pt.y);
                    animators.lhc_b1 = requestAnimationFrame(orbit);
                }
                animators.lhc_b1 = requestAnimationFrame(orbit);
            });

            cancelAnimationFrame(animators.lhc_b2);
            animators.lhc_b2 = requestAnimationFrame((ts) => {
                let lhc_progress = 1 - (((ts - start_time) % speed_b2) / speed_b2);
                let pt = getPointOnPath(path_lhc_b2, lhc_progress);
                p_lhc_b2.setAttribute("cx", pt.x);
                p_lhc_b2.setAttribute("cy", pt.y);
                
                // Keep orbiting CCW
                function orbit(t) {
                    let p = 1 - ((t % speed_b2) / speed_b2);
                    let pt = getPointOnPath(path_lhc_b2, p);
                    p_lhc_b2.setAttribute("cx", pt.x);
                    p_lhc_b2.setAttribute("cy", pt.y);
                    animators.lhc_b2 = requestAnimationFrame(orbit);
                }
                animators.lhc_b2 = requestAnimationFrame(orbit);
            });

            if (progress < 1.0) {
                requestAnimationFrame(ramp);
            } else {
                updateSystemStatus("LHC_STABLE", "STABLE BEAMS (6.8 TeV)");
                btn_collide.classList.remove("btn-disabled");
                node_psb.classList.remove("svg-node-active");
                node_ps.classList.remove("svg-node-active");
                node_sps.classList.remove("svg-node-active");
                det_atlas.classList.add("svg-node-active");
                det_cms.classList.add("svg-node-active");
                det_alice.classList.add("svg-node-active");
                det_lhcb.classList.add("svg-node-active");
            }
        }

        requestAnimationFrame(ramp);
    });

    // Physics update readout
    function updateDashboardPhysics(energy) {
        val_energy.innerText = (energy / 1000).toFixed(2) + " TeV";
        
        // B field = p / (0.3 * R)
        let B_field = energy / (0.299792458 * bending_radius);
        val_field.innerText = B_field.toFixed(3) + " T";

        // Lorentz Gamma
        let gamma = energy / proton_mass;
        val_gamma.innerText = gamma.toLocaleString('de-DE', {maximumFractionDigits: 0});

        // Luminosity scale
        let lumi = 0.0;
        if (system_state === "LHC_STABLE") {
            lumi = 2.1e34;
        } else if (system_state === "LHC_RAMPING") {
            let progress = (energy - 450) / (6800 - 450);
            lumi = progress * 1.5e34;
        }
        val_lumi.innerText = lumi > 0 ? (lumi / 1e34).toFixed(2) + " x10³⁴" : "0.00";
    }

    // 7. COLLIDE
    btn_collide.addEventListener("click", () => {
        collision_count += 5;
        collision_counter_text.innerText = "Kollisionen: " + collision_count;

        // Visual flash in SVG
        let selected_node = det_atlas;
        if (selected_detector === "CMS") selected_node = det_cms;
        if (selected_detector === "ALICE") selected_node = det_alice;
        if (selected_detector === "LHCb") selected_node = det_lhcb;

        selected_node.classList.add("svg-node-active-red");
        setTimeout(() => {
            selected_node.classList.remove("svg-node-active-red");
        }, 300);

        // Simulated Collision Tracks
        triggerCollisionEventDisplay();
        
        // Add fake events to mass spectrum
        generateEventsForHistogram();
        drawHistogram();
    });

    // --------------------------------------------------------------------------
    // Event Display Drawing
    // --------------------------------------------------------------------------
    function drawStaticDetectorBackground() {
        ctx_event.clearRect(0, 0, canvas_event.width, canvas_event.height);
        
        // Dimensions
        const cx = canvas_event.width / 2;
        const cy = canvas_event.height / 2;
        
        // Draw concentric circles representing detector layers (Silicon tracker, EM Calo, Hadron Calo, Muon chambers)
        ctx_event.strokeStyle = "#21262d";
        ctx_event.lineWidth = 1;
        
        ctx_event.beginPath(); ctx_event.arc(cx, cy, 20, 0, 2*Math.PI); ctx_event.stroke(); // Inner Tracker
        ctx_event.beginPath(); ctx_event.arc(cx, cy, 45, 0, 2*Math.PI); ctx_event.stroke(); // ECAL
        ctx_event.beginPath(); ctx_event.arc(cx, cy, 70, 0, 2*Math.PI); ctx_event.stroke(); // HCAL
        ctx_event.strokeStyle = "#30363d";
        ctx_event.beginPath(); ctx_event.arc(cx, cy, 95, 0, 2*Math.PI); ctx_event.stroke(); // Muon Chambers

        // Text labels inside the canvas
        ctx_event.fillStyle = "#8b949e";
        ctx_event.font = "10px monospace";
        ctx_event.fillText(selected_detector + " EVENT MONITOR", 10, 20);
        ctx_event.fillText("STATUS: ACTIVE", 10, 35);
    }

    function triggerCollisionEventDisplay() {
        drawStaticDetectorBackground();
        const cx = canvas_event.width / 2;
        const cy = canvas_event.height / 2;

        const num_tracks = 12 + Math.floor(Math.random() * 10);
        
        // Draw beautiful animated explosion lines (Tracks)
        for (let i = 0; i < num_tracks; i++) {
            const angle = Math.random() * 2 * Math.PI;
            const length = 40 + Math.random() * 55;
            const is_muon = Math.random() > 0.7; // Muons reach the outer edge
            
            ctx_event.beginPath();
            ctx_event.moveTo(cx, cy);
            
            // Slightly curved paths representing particles in magnetic field
            const c_angle = angle + (Math.random() > 0.5 ? 0.15 : -0.15);
            const target_x = cx + length * Math.cos(angle);
            const target_y = cy + length * Math.sin(angle);
            const cp_x = cx + (length / 2) * Math.cos(c_angle);
            const cp_y = cy + (length / 2) * Math.sin(c_angle);

            ctx_event.quadraticCurveTo(cp_x, cp_y, target_x, target_y);
            
            // Styles for different particle types
            if (is_muon) {
                ctx_event.strokeStyle = "#2ea44f"; // Green Muon track
                ctx_event.lineWidth = 2;
                ctx_event.stroke();
                // Muon chamber hit
                ctx_event.fillStyle = "#2ea44f";
                ctx_event.beginPath();
                ctx_event.arc(target_x, target_y, 3, 0, 2*Math.PI);
                ctx_event.fill();
            } else {
                ctx_event.strokeStyle = Math.random() > 0.5 ? "#58a6ff" : "#ff7f0e"; // Electrons/Pions
                ctx_event.lineWidth = 1;
                ctx_event.stroke();
                // Calorimeter shower
                ctx_event.fillStyle = "rgba(255, 215, 0, 0.4)";
                ctx_event.beginPath();
                ctx_event.arc(target_x, target_y, 2 + Math.random()*3, 0, 2*Math.PI);
                ctx_event.fill();
            }
        }
    }

    // --------------------------------------------------------------------------
    // Histogram Spektrum Reconstruction
    // --------------------------------------------------------------------------
    function generateEventsForHistogram() {
        // Accumulate invariant mass samples
        // We simulate a background (exponential) + Z0 peak (91 GeV) + Higgs peak (125 GeV)
        // More collisions = peak stands out
        for (let i = 0; i < 40; i++) {
            // Background: Exponentially falling
            let mass = 50 + Math.exp(Math.random() * 4.6) * 1.5; // Spread mostly in lower range
            if (mass > 50 && mass < 150) {
                accumulated_mass_data.push(mass);
            }
        }
        
        // Z0 Peak (91 GeV) - High cross section
        for (let i = 0; i < 15; i++) {
            let mass = 91.2 + (Math.random() - 0.5) * 4.0 + (Math.random() - 0.5) * 2.0;
            accumulated_mass_data.push(mass);
        }

        // Higgs Peak (125 GeV) - Low cross section (rarer, but emerges)
        for (let i = 0; i < 4; i++) {
            let mass = 125.0 + (Math.random() - 0.5) * 3.0 + (Math.random() - 0.5) * 1.5;
            accumulated_mass_data.push(mass);
        }
    }

    function drawHistogram() {
        ctx_hist.clearRect(0, 0, canvas_hist.width, canvas_hist.height);
        
        const w = canvas_hist.width;
        const h = canvas_hist.height;

        // Draw axes lines
        ctx_hist.strokeStyle = "#30363d";
        ctx_hist.lineWidth = 1;
        ctx_hist.beginPath();
        ctx_hist.moveTo(30, 10);
        ctx_hist.lineTo(30, h - 20);
        ctx_hist.lineTo(w - 10, h - 20);
        ctx_hist.stroke();

        // X Axis labels (50 to 150 GeV)
        ctx_hist.fillStyle = "#8b949e";
        ctx_hist.font = "8px sans-serif";
        ctx_hist.fillText("50 GeV/c²", 30, h - 8);
        ctx_hist.fillText("100 GeV/c²", w/2 - 10, h - 8);
        ctx_hist.fillText("150 GeV/c²", w - 45, h - 8);

        if (accumulated_mass_data.length === 0) {
            ctx_hist.fillStyle = "#8b949e";
            ctx_hist.font = "11px monospace";
            ctx_hist.fillText("NO DATA. RUN COLLISIONS TO ACCUMULATE EVENTS", w/2 - 110, h/2);
            return;
        }

        // Bin data (30 bins between 50 and 150 GeV)
        const min_mass = 50;
        const max_mass = 150;
        const num_bins = 40;
        const bins = new Array(num_bins).fill(0);

        accumulated_mass_data.forEach(val => {
            if (val >= min_mass && val < max_mass) {
                let bin_idx = Math.floor(((val - min_mass) / (max_mass - min_mass)) * num_bins);
                if (bin_idx >= 0 && bin_idx < num_bins) {
                    bins[bin_idx]++;
                }
            }
        });

        // Find max bin value to scale the Y axis
        const max_count = Math.max(...bins, 1);

        // Draw bins (Bars)
        const bar_w = (w - 45) / num_bins;
        for (let i = 0; i < num_bins; i++) {
            const bar_h = (bins[i] / max_count) * (h - 40);
            const x = 30 + i * bar_w;
            const y = h - 20 - bar_h;

            ctx_hist.fillStyle = "rgba(88, 166, 255, 0.4)";
            ctx_hist.fillRect(x, y, bar_w - 1, bar_h);
            
            // Neon top line
            ctx_hist.fillStyle = "#58a6ff";
            ctx_hist.fillRect(x, y, bar_w - 1, 2);
        }

        // Draw a soft overlay curve showing the fitted Breit-Wigner at Z0 and Higgs (once many events accumulate)
        if (collision_count > 10) {
            ctx_hist.strokeStyle = "rgba(248, 81, 73, 0.6)";
            ctx_hist.lineWidth = 1.5;
            ctx_hist.beginPath();
            
            for (let x_pix = 30; x_pix < w - 10; x_pix++) {
                let mass_val = min_mass + ((x_pix - 30) / (w - 40)) * (max_mass - min_mass);
                // Background exponential
                let bg = Math.exp(-(mass_val - 50) / 25) * 18;
                // Z0 Resonance
                let z_peak = (1 / ((mass_val - 91.2)**2 + 2.5**2)) * 320;
                // Higgs Resonance
                let higgs_peak = (1 / ((mass_val - 125)**2 + 1.6**2)) * 12;
                
                let y_val = bg + z_peak + higgs_peak;
                // Scale y_val to fit screen
                let y_pix = h - 20 - (y_val / 20) * (h - 45);
                y_pix = Math.max(10, Math.min(h-20, y_pix));

                if (x_pix === 30) {
                    ctx_hist.moveTo(x_pix, y_pix);
                } else {
                    ctx_hist.lineTo(x_pix, y_pix);
                }
            }
            ctx_hist.stroke();
            
            // Labels for Z and Higgs
            ctx_hist.fillStyle = "#ff7f0e";
            ctx_hist.fillText("Z⁰ (91 GeV)", w * 0.43, 25);
            
            ctx_hist.fillStyle = "#2ea44f";
            ctx_hist.fillText("Higgs (125 GeV)", w * 0.70, 45);
        }
    }
})();
</script>
'''))
```

## 📈 Post-Kollisions-Analyse

Nachdem hunderte von Kollisionen in der Schaltzentrale ausgeführt wurden, sammeln die Detektoren Millionen von Ereignissen. Nun nutzen wir Python, um eine präzise wissenschaftliche Analyse der invarianten Massen durchzuführen. Hier simulieren wir die Datenaufnahme und fitten eine Kombination aus einem exponentiellen Untergrund (durch unkorrelierte Teilchenjets) und zwei Breit-Wigner-Resonanzen für das **Z0-Boson** und das **Higgs-Boson**.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 1. Erzeugung realistischer Detektordaten
np.random.seed(137) # Physikalische Konsistenz
n_events = 25000

# Exponentieller kombinatorischer Untergrund
background = np.random.exponential(scale=35.0, size=int(0.75 * n_events)) + 50.0
background = background[background < 160.0]

# Z0 Resonanz (M = 91.18 GeV, Breit-Wigner mit Gauß-Detektorauflösung verschmiert)
z_signal = np.random.normal(loc=91.18, scale=2.8, size=int(0.23 * n_events))
z_signal = z_signal[(z_signal >= 50.0) & (z_signal <= 160.0)]

# Higgs-Signal (M = 125.1 GeV, sehr kleiner Wirkungsquerschnitt, ca. 2% aller Ereignisse)
higgs_signal = np.random.normal(loc=125.1, scale=1.8, size=int(0.02 * n_events))
higgs_signal = higgs_signal[(higgs_signal >= 50.0) & (higgs_signal <= 160.0)]

# Gesamtdaten verschmelzen
all_masses = np.concatenate([background, z_signal, higgs_signal])

# 2. Histogramm-Eigenschaften definieren
bins = 80
bin_heights, bin_edges = np.histogram(all_masses, bins=bins, range=(50, 160))
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])
bin_width = bin_edges[1] - bin_edges[0]

# 3. Fitfunktion definieren: Untergrund (Exp) + Z-Peak (Breit-Wigner) + Higgs-Peak (Gauß)
def fit_model(x, a0, lambda_bg, a_z, mean_z, gamma_z, a_h, mean_h, sigma_h):
    # Exponentieller Untergrund
    bg = a0 * np.exp(-(x - 50) / lambda_bg)
    # Relativistische Breit-Wigner für das Z-Boson
    z_peak = a_z * (gamma_z / 2)**2 / ((x - mean_z)**2 + (gamma_z / 2)**2)
    # Gauß für das Higgs-Boson (dominert von Detektorauflösung)
    higgs_peak = a_h * np.exp(-0.5 * ((x - mean_h) / sigma_h)**2)
    return bg + z_peak + higgs_peak

# Startwerte für den Fit-Algorithmus
initial_guess = [
    500.0, 30.0,      # Untergrund-Parameter
    250.0, 91.2, 5.0, # Z0-Boson Parameter (Breit-Wigner)
    25.0, 125.0, 2.0  # Higgs-Boson Parameter (Gauß)
]

popt, pcov = curve_fit(fit_model, bin_centers, bin_heights, p0=initial_guess)

# 4. Plotten der wissenschaftlichen Ergebnisse
cu.apply_cern_style() if 'cu' in globals() else plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 7))

# Datenpunkte mit Poisson-Statistik-Fehlerbalken
errors = np.sqrt(bin_heights)
ax.errorbar(bin_centers, bin_heights, yerr=errors, fmt='o', color='#8b949e', 
            markersize=4, elinewidth=1, capsize=1.5, label='Kollisionsdaten (CERN 2026 Run)')

# Fit-Kurve einzeichnen
x_fit = np.linspace(50, 160, 1000)
y_fit = fit_model(x_fit, *popt)
ax.plot(x_fit, y_fit, color='#58a6ff', linewidth=2.5, label='Gesamt-Fit (Untergrund + Peaks)')

# Einzelkomponenten des Fits zur Veranschaulichung einzeichnen
y_bg = popt[0] * np.exp(-(x_fit - 50) / popt[1])
y_z = popt[2] * (popt[4] / 2)**2 / ((x_fit - popt[3])**2 + (popt[4] / 2)**2)
y_h = popt[5] * np.exp(-0.5 * ((x_fit - popt[6]) / popt[7])**2)

ax.plot(x_fit, y_bg, '--', color='#8b949e', alpha=0.7, label='Kombinatorischer Untergrund')
ax.plot(x_fit, y_z + y_bg, ':', color='#ff7f0e', label='Z⁰ Signalüberhöhung')
ax.plot(x_fit, y_h + y_bg, ':', color='#2ea44f', label='Higgs Signalüberhöhung')

# Beschriftungen und Styling
ax.set_title('⚛️ LHC REKONSTRUIERTES Z⁰ & HIGGS BOSON MASSENSPEKTRUM', fontsize=14, fontweight='bold', color='#58a6ff')
ax.set_xlabel('Invariante Rekonstruierte Masse $M_{\\mu\\mu}$ bzw. $M_{\\gamma\\gamma}$ [GeV/c²]', fontsize=12, color='#e6edf3')
ax.set_ylabel(f'Ereignisse pro {bin_width:.2f} GeV/c²', fontsize=12, color='#e6edf3')
ax.set_xlim(50, 160)
ax.set_ylim(0, np.max(bin_heights) * 1.15)
ax.legend(framealpha=0.8, loc='upper right')

# Textboxen für physikalische Eigenschaften
textstr = '\n'.join((
    r'$M(Z^0) = %.2f \\pm %.2f \\text{ GeV/c}^2$' % (popt[3], np.sqrt(pcov[3,3])),
    r'$\Gamma(Z^0) = %.2f \\text{ GeV}$' % (popt[4],),
    r'$M(H) = %.2f \\pm %.2f \\text{ GeV/c}^2$' % (popt[6], np.sqrt(pcov[6,6]))
))
props = dict(boxstyle='round', facecolor='#161b22', edgecolor='#30363d', alpha=0.8)
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props, color='#e6edf3')

plt.tight_layout()
plt.show()
```
