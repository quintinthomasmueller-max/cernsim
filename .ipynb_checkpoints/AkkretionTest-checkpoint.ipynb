## Sternentstehung

### Virialsatz und Akkretionsscheiben

Wir betrachten einen Ring mit Radius $r$ und Masse $m$ um eine zentrale Masse $M$, die ein schwarzes Loch sein könnte. Der Ring ist Teil einer Scheibe. Der nach außen anschließende Ring läuft etwas langsamer um die Scheibe; der nach innen anschließende etwas schneller. Dadurch entsteht Reibung: Die Scheibe heizt sich auf. Zugleich verliert jeder Ring Energie und bewegt sich langsam auf die zentrale Masse zu.

Der Radius $r$ schrumpft um $\mathrm{d}r < 0$. Dadurch ändert sich seine potentielle Energie um
$$\mathrm{d}E_\mathrm{pot} = \frac{\mathrm{d}E_\mathrm{pot}}{\mathrm{d}r}\mathrm{d}r =
  \frac{GMm}{r^2}\mathrm{d}r\;.$$
Aufgrund des Virialsatzes wird die Hälfte davon in kinetische Energie umgesetzt; die andere Hälfte wird abgestrahlt:
$$\mathrm{d}E_\mathrm{S} = -\frac{1}{2}\mathrm{d}E_\mathrm{pot} = -\frac{GMm}{2r^2}\mathrm{d}r\;.$$
Wir setzen hier $\mathrm{d}r = -|\mathrm{d}r|$ ein, um nicht mehr darauf achten zu müssen, dass $\mathrm{d}r$ negativ ist. Die pro Zeiteinheit abgestrahlte Energie ist die Leuchtkraft,
$$\mathrm{d}L = \frac{\mathrm{d}E_\mathrm{S}}{\mathrm{d}t} =
  \frac{GM\dot m}{2r^2}|\mathrm{d}r|\;,$$
wobei $\dot m$ die Masse ist, die pro Zeiteinheit zur zentralen Masse hin strömt. Sie wird auch als *Akkretionsrate* bezeichnet.

Nach dem Stefan-Boltzmann'schen Gesetz ist die von einer Fläche $\mathrm{d}A$ bei der Temperatur $T$ abgestrahlte Leuchtkraft durch
$$\mathrm{d}L = \sigma T^4\mathrm{d}A$$
gegeben, wobei die Stefan-Boltzmann-Konstante
$$\sigma = 5.6704\cdot10^{-8}\,\frac{\mathrm{W}}{\mathrm{m}^2\mathrm{K}^4}$$
auftritt. Die Fläche eines Kreisrings mit Radius $r$ und Dicke $|\mathrm{d}r|$ ist
$$\mathrm{d}A = 2\cdot2\pi r|\mathrm{d}r|\;,$$
wobei der Faktor $2$ berücksichtigt, dass der Ring eine Ober- und eine Unterseite hat. Also ist
$$\mathrm{d}L = 4\pi\sigma T^4r|\mathrm{d}r|\;.$$

Diesen Ausdruck für $\mathrm{d}L$ können wir dem vorigen Ergebnis aus dem Virialsatz gleichsetzen und erhalten
$$4\pi\sigma T^4r = \frac{GM\dot m}{2r^2}\;.$$
Für die Temperatur am Radius $r$ einer Akkretionsscheibe bekommen wir also das Ergebnis
$$T(r) = \left(\frac{GM\dot m}{8\pi r^3\sigma}\right)^{1/4}\;.$$

Indem wir hier den Schwarzschild-Radius
$$r_\mathrm{S} = \frac{2GM}{c^2}$$
der zentralen Masse einsetzen und etwas umformen, erhalten wir
$$T(r) = \left(\frac{c^6}{64\pi\sigma G^2}\right)^{1/4}\,
  \left(\frac{r_\mathrm{S}}{r}\right)^{3/4}\frac{\dot m^{1/4}}{M^{1/2}}\;.$$
Indem wir die Konstanten einsetzen, erhalten wir
$$T(r) = 3.88\cdot10^6\,\mathrm{K}\left(\frac{M}{10^6\,M_\odot}\right)^{-1/2}
  \left(\frac{\dot m}{M_\odot\,\mathrm{yr}^{-1}}\right)^{1/4}
  \left(\frac{r}{r_\mathrm{S}}\right)^{-3/4}\;.$$
  
Akkretionsscheiben um schwarze Löcher sind also umso heißer, je weniger Masse das schwarze Loch hat. Die Temperatur steigt nach innen an, und sie nimmt mit der Akkretionsrate zu.

### Vorbereitungen

```python
import numpy as np
import matplotlib.pyplot as pl
from scipy.optimize import root_scalar
from collections import namedtuple

cmap = pl.get_cmap("tab10")

pl.rcParams.update({"figure.figsize": (12, 9),
                     "font.size": 12,
                     "font.family": "serif", "font.serif": "Dejavu Serif",
                     "mathtext.fontset": "dejavuserif"})

# Strukturierte Konstanten mit namedtuple für sauberen Zugriff
Constants = namedtuple('Constants', ['c', 'k_B', 'G_N', 'h_bar', 'm_Sun', 'yr', 'L_Sun', 'sigma'])
cst = Constants(
    c=2.9979e10,      # Lichtgeschwindigkeit [cm/s]
    k_B=1.3806e-16,    # Boltzmann-Konstante [erg/K]
    G_N=6.6743e-8,     # Gravitationskonstante [cm^3/g/s^2]
    h_bar=1.0546e-27,  # Reduziertes Planck-Wirkungsquantum [erg s]
    m_Sun=1.9884e33,   # Sonnenmasse [g]
    yr=3.1558e7,       # Jahr [s]
    L_Sun=3.8460e33,   # Sonnenleuchtkraft [erg/s]
    sigma=5.6704e-5    # Stefan-Boltzmann-Konstante [erg/cm^2/s/K^4]
)

## Numerische Konstante C für das Temperaturprofil
# Aufteilung zur besseren Nachvollziehbarkeit
C_pre = cst.c**6 / (64 * np.pi * cst.sigma * cst.G_N**2)
C = (C_pre / (cst.m_Sun / cst.yr))**0.25 / 1e3

x_max = root_scalar(lambda x: (3-x)*np.exp(x)-3, bracket=[1,10]).root

def r_s(m):
    """
    Berechnet den Schwarzschild-Radius in cm.
    
    Parameters:
    m (float/array): Zentrale Masse in Millionen Sonnenmassen.
    """
    return 2 * cst.G_N * (m * 1e6 * cst.m_Sun) / cst.c**2

def temperature(r, m, mdot):
    """
    Berechnet die Temperatur in K. Unterstützt Vektorisierung.
    
    Parameters:
    r (float/array): Radius in Schwarzschild-Radien.
    m (float): Zentrale Masse in Millionen Sonnenmassen.
    mdot (float): Akkretionsrate in Sonnenmassen pro Jahr.
    """
    return C * mdot**0.25 / np.sqrt(m) / r**0.75

def u_planck(T):
    """
    Energiedichte des Planck-Spektrums in erg/cm^3 (vektorisiert).
    """
    return np.pi**2 / 15 * (cst.k_B * T)**4 / (cst.h_bar * cst.c)**3

def du_planck(x):
    """
    Dimensionsloses Planck-Spektrum (vektorisiert).
    
    Parameters:
    x (float/array): Frequenz in Einheiten von k_BT/h.
    """
    # Sicherheitscheck für x=0 (Vermeidung von Division durch Null)
    val = np.zeros_like(x, dtype=float)
    mask = x > 0
    e = np.exp(-x[mask])
    val[mask] = x[mask]**3 * e / (1 - e)
    return val

def nu_max(T):
    """
    Frequenz maximaler Intensität nach Wien'schem Verschiebungsgesetz in Hz.
    """
    return x_max * cst.k_B / (2 * np.pi * cst.h_bar) * T

def wavelength(nu):
    """
    Wellenlänge in nm für gegebene Frequenz in Hz.
    """
    return 1e9 * cst.c / nu

def frequency(la):
    """
    Frequenz in Hz für gegebene Wellenlänge in nm.
    """
    return cst.c / (la / 1e7)
```

### Temperatur als Funktion des Radius für verschiedene zentrale Massen

```python
r = np.logspace (log (3)/log (10), 3, 128)
m    = [0.1, 1, 10]
mdot = [0.5, 1, 2]
lt   = ['--', '-', ':']
for i in range (0, len (m)):
    for j in range (0, len (mdot)):
        pl.plot (r, [temperature (v, m[i], mdot[j]) for v in r], ls = lt[j],
                 label = "$M = "+
                 "{:3.1f}".format (m[i])+"\\cdot10^6\\,M_\\odot$, $\\dot m = "+
                 "{:3.1f}".format (mdot[j])+"\\,M_\\odot\\mathrm{yr}^{-1}$",
                 color = cmap (i))
pl.legend (loc = "best")
pl.xscale ("log")
pl.yscale ("log")
pl.xlabel ("Radius $r$ in Schwarzschild-Radien")
pl.ylabel ("Temperatur $T$ in K")
pl.grid ()
pl.show ()
```

```python
x = np.logspace (-1, 1, 128)
T = [2500, 5000, 7500, 10000]
for i in range (0, len (T)):
    nu0 = cst['k_B']*T[i]/(2*np.pi*cst['h_bar'])
    u   = u_planck (T[i])
    pl.plot ([v*nu0 for v in x], [u*du_planck (v)/nu0 for v in x],
             label = "$T = "+"{:4.0f}".format (T[i])+"$",
             color = cmap (i+1))
    pl.plot ([x_max*nu0], [u*du_planck (x_max)/nu0], "o", color = cmap (i+1))
pl.legend (loc = 'best')
pl.xscale ('log')
pl.yscale ('log')
y0, y1 = pl.ylim ()
pl.fill_betweenx ([y0, y1], frequency (350), frequency (800), color = cmap (0), alpha = 0.25)
pl.xlabel ('Frequenz $\\nu$ in Hz')
pl.ylabel ('Strahlungsintensität in erg/s/cm$^3$/Hz')
pl.grid ()
pl.show ()
```

```python
from scipy.integrate import quad

def spectrum(x):
    # Vectorized integration is tricky with quad, we can keep quad but vectorize outside
    integral, _ = quad(lambda r_r: (r_r/3)**0.75/r_r**2 * du_planck(x * (r_r/3)**0.75), 3, np.inf)
    return 45/np.pi**4 * integral

# Vectorize the spectrum function
spectrum_vec = np.vectorize(spectrum)

def luminosity(mdot):
    return mdot/12 * cst.m_Sun / cst.yr * cst.c**2 / cst.L_Sun

m_list    = [1, 10, 100]
mdot_list = [0.5, 1]
lt        = ['--', '-', ':']

x = np.logspace(-1, 1, 128)
pl.figure()
for i, m_val in enumerate(m_list):
    T_3 = temperature(3, m_val, 1) # temp at r=3
    nu0 = cst.k_B * T_3 / (2 * np.pi * cst.h_bar)
    for j, mdot_val in enumerate(mdot_list):
        L = luminosity(mdot_val)
        nu_array = nu0 * x
        L_nu = L * spectrum_vec(x)
        pl.plot(nu_array, L_nu, ls=lt[j], color=cmap(i+1),
                 label=f"$M = {m_val}\\times 10^6 M_\\odot$, $\\dot m = {mdot_val} M_\\odot\\mathrm{{yr}}^{{-1}}$")

pl.xscale('log')
pl.yscale('log')
pl.xlabel('Frequenz $\\nu$ in Hz')
pl.ylabel('spezifische Leuchtkraft in $L_\\odot$')
pl.legend(loc='best', fontsize=10)
y0, y1 = pl.ylim()
pl.fill_betweenx([y0, y1], frequency(800), frequency(350), color=cmap(0), alpha=0.25, label='Sichtbarer Bereich')
pl.grid(True, which="both", ls="-", alpha=0.5)
pl.show()
```
